```python
import pandas as pd
import numpy as np
from pathlib import Path

# Get the directory where the current notebook is located
current_dir = Path.cwd()

# Define the path relative to the notebook
data_path = current_dir.parent / "data" / "epidemiology.csv"
```

```python
# Load the dataset
data=pd.read_csv(data_path)
data.head(2)
```

```python
# Get unique values from the first column (column index 0)
unique_first_cells = data["location_key"].dropna().unique()
print(unique_first_cells)
```

```python
# Return unique location keys
def get_unique_location_keys(data):
    return list(set(data["location_key"].dropna().str.split("_").str[0]))

unique_Country_Keys = get_unique_location_keys(data)

print(unique_Country_Keys)
print(len(unique_Country_Keys))
```

```python
focus_data = data[data['location_key'].isin(unique_Country_Keys)]
focus_data.info()
```
